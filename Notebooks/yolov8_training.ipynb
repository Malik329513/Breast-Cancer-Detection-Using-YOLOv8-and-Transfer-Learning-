# *Object Detection Prediction*

```python
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display

# Load a single image for visualization (replace with the actual image path)
image = Image.open('/kaggle/input/yolov007/a5/train/images/P_00027_png.rf.44a24a67bcfecc607930dbeceee979b6.jpg')

# Perform prediction on the image
result = model.predict(source=image)

# Draw the results on the image
draw = ImageDraw.Draw(image)
font = ImageFont.load_default()

# Mapping from ID to class name
id_to_name = {0: 'BENIGN', 1: 'BENIGN_WITHOUT_CALLBACK', 2: 'MALIGNANT'}

# Iterate over each detected object
for i, box in enumerate(result[0].boxes.xyxy):
    box = box.tolist()  # Convert to list
    box = [int(coord) for coord in box]  # Ensure coordinates are integers

    class_id = int(result[0].boxes.cls[i])
    class_name = id_to_name[class_id]
    confidence = result[0].boxes.conf[i]

    # Draw the bounding box and label
    draw.rectangle(box, outline="red", width=3)
    draw.text((box[0], box[1]), f"{class_name} {confidence:.2f}", fill="white", font=font)

# Save the image
output_path = '/kaggle/working/output_image.png'
image.save(output_path)

# Display the image inline
display(image)
```

# Comprehensive Performance Evaluation of YOLOv8 Model on Breast Cancer Detection

```python
import os
from ultralytics import YOLO
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Paths to image and annotation folders
image_folder = '/kaggle/input/yolov007/a5/train/images'
annotations_folder = '/kaggle/input/yolov007/a5/train/labels'

# Load the YOLO model
model = YOLO('/kaggle/working/runs/detect/train2/weights/best.pt')

# Mapping class indices to label names
label_map = {0: 'BENIGN', 1: 'BENIGN_WITHOUT_CALLBACK', 2: 'MALIGNANT'}

# Get all image paths from the folder
def get_image_paths(folder_path):
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff'}
    return [os.path.join(folder_path, f) for f in os.listdir(folder_path) if os.path.splitext(f)[1].lower() in image_extensions]

image_paths = get_image_paths(image_folder)

# Function to get true labels from the annotation file
def get_true_labels_from_annotation(image_path):
    filename = os.path.basename(image_path).replace('.jpg', '.txt').replace('.png', '.txt')
    annotation_path = os.path.join(annotations_folder, filename)

    true_labels = []
    if os.path.exists(annotation_path):
        with open(annotation_path, 'r') as file:
            lines = file.readlines()
            for line in lines:
                class_id = int(line.split()[0])
                true_labels.append(class_id)
    return true_labels

# Store true and predicted labels
true_labels_list = []
predicted_labels_list = []
predicted_scores_list = []

# Loop through all images and process predictions
for image_path in image_paths:
    # Get true labels from annotation file
    true_labels = get_true_labels_from_annotation(image_path)

    # Perform prediction on the image
    results = model.predict(image_path)

    # Extract the predicted labels
    predicted_labels = []
    if results[0].boxes:
        for box in results[0].boxes:
            predicted_labels.append(int(box.cls))


    for result in results:
        if result.boxes is not None:
            confidences = result.boxes.conf  # Confidence scores
            num_boxes = len(confidences)  # Number of detected boxes
            print(f"Number of boxes detected: {num_boxes}")  # Debug statement to show number of detections
            predicted_scores_list.extend(confidences.tolist())


    # Add to the main lists, ensuring labels are aligned
    true_labels_list.extend(true_labels)
    predicted_labels_list.extend(predicted_labels)


# Ensure that true_labels_list and predicted_labels_list have the same length
if len(true_labels_list) != len(predicted_labels_list):
    # Print diagnostic information
    print(f"Length mismatch: true_labels ({len(true_labels_list)}) vs. predicted_labels ({len(predicted_labels_list)})")
    print(f"True labels: {true_labels_list}")
    print(f"Predicted labels: {predicted_labels_list}")

    # Handle length mismatch: truncate the longer list or handle separately if needed
    min_length = min(len(true_labels_list), len(predicted_labels_list))
    true_labels_list = true_labels_list[:min_length]
    predicted_labels_list = predicted_labels_list[:min_length]
    predicted_scores_list = predicted_scores_list[:min_length]

# Compute confusion matrix
cm = confusion_matrix(true_labels_list, predicted_labels_list, labels=[0, 1, 2])

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['BENIGN', 'BENIGN_WITHOUT_CALLBACK', 'MALIGNANT'])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

# Example: Replace these with your actual lists
# y_true = [0, 2, 0, 2, 0, 2]  # Replace with your true labels
# y_pred = [0, 0, 0, 2, 2, 2]  # Replace with your predicted labels

# Convert lists to numpy arrays
y_true = np.array(true_labels_list)
y_pred = np.array(predicted_labels_list)

# Compute precision, recall, and F1-score for each class
precision, recall, fscore, support = precision_recall_fscore_support(y_true, y_pred, average=None, labels=[0, 2])

# Compute overall metrics
overall_precision, overall_recall, overall_fscore, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
overall_accuracy = accuracy_score(y_true, y_pred)

# Print the overall metrics
print(f'Overall Accuracy: {overall_accuracy:.4f}')
print(f'Overall Precision: {overall_precision:.4f}')
print(f'Overall Recall: {overall_recall:.4f}')
print(f'Overall F1 Score: {overall_fscore:.4f}')

# Class labels
class_labels = ['BENIGN', 'MALIGNANT']

# Plot Precision, Recall, and F-score
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Precision
ax[0].bar(class_labels, precision, color='blue', alpha=0.7)
ax[0].set_title('Precision per Class')
ax[0].set_xlabel('Classes')
ax[0].set_ylabel('Precision')

# Recall
ax[1].bar(class_labels, recall, color='green', alpha=0.7)
ax[1].set_title('Recall per Class')
ax[1].set_xlabel('Classes')
ax[1].set_ylabel('Recall')

# F-score
ax[2].bar(class_labels, fscore, color='red', alpha=0.7)
ax[2].set_title('F-score per Class')
ax[2].set_xlabel('Classes')
ax[2].set_ylabel('F-score')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# Example true and predicted labels



# Convert lists to numpy arrays
y_true = np.array(true_labels_list)
y_pred = np.array(predicted_labels_list)
y_scores = np.array(predicted_scores_list)

# Compute precision and recall for each class
class_labels = [0, 2]  # Update according to your class labels

precision = {}
recall = {}
thresholds = {}

for class_label in class_labels:
    # Binarize the true labels and predicted scores for the specific class
    y_true_bin = (y_true == class_label).astype(int)
    y_scores_bin = np.array([score if pred == class_label else 0 for pred, score in zip(y_pred, y_scores)])

    # Compute precision and recall for each threshold
    precision[class_label], recall[class_label], thresholds[class_label] = precision_recall_curve(y_true_bin, y_scores_bin)

    # Include thresholds for plotting convenience (extend with the highest threshold)
    thresholds[class_label] = np.concatenate(([0], thresholds[class_label]))

# Plot Precision and Recall as line graphs
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot Precision
for class_label in class_labels:
    ax[0].plot(thresholds[class_label], precision[class_label], label=f'Class {class_label}')
ax[0].set_title('Precision vs Threshold')
ax[0].set_xlabel('Threshold')
ax[0].set_ylabel('Precision')
ax[0].legend()

# Plot Recall
for class_label in class_labels:
    ax[1].plot(thresholds[class_label], recall[class_label], label=f'Class {class_label}')
ax[1].set_title('Recall vs Threshold')
ax[1].set_xlabel('Threshold')
ax[1].set_ylabel('Recall')
ax[1].legend()

plt.tight_layout()
plt.show()
```
